# **Linear Regression**

**Load the dataset**


```python
import pandas as pd

df = pd.read_csv('Ecommerce Customers')
df.head()
```

**Explore the data**

```python
df.head()
df.info()
df.describe()
```

**Data Cleaning**

```python
df.isnull().sum()
```

No missing values found, no cleaning needed.

**Feature Engineering**

```python
df.columns
```

```python
X = df[['Avg. Session Length', 'Time on App',
        'Time on Website', 'Length of Membership']]

y = df['Yearly Amount Spent']
```

**Prepare data for modeling**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

**Train the model**

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

**Evaluate the model**

**Predictions**



```python
predictions = model.predict(X_test)
```

**Metrics**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
coeff_df = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```

The Linear Regression model provides a good fit for the dataset.

- Predictions are closely aligned with actual values.
- The scatter plot confirms a strong linear relationship.
